```python
#Change this to your files/directories
dataset_dir = '/kaggle/input/ethical-unethical/Ethical-Unethical-Dataset.csv'
model_names=['tasksource/deberta-small-long-nli',
            'facebook/bart-large-mnli',
             'MoritzLaurer/DeBERTa-v3-xsmall-mnli-fever-anli-ling-binary','MoritzLaurer/DeBERTa-v3-base-mnli-fever-anli',"MoritzLaurer/DeBERTa-v3-large-mnli-fever-anli-ling-wanli", "MoritzLaurer/deberta-v3-large-zeroshot-v2.0-c",
             "mlburnham/Political_DEBATE_base_v1.0",'mlburnham/Political_DEBATE_large_v1.0']
```

```python
import csv
from transformers import pipeline, AutoModel, AutoTokenizer
from tqdm import tqdm
import statistics

with open(dataset_dir, newline='') as f:
    reader = csv.reader(f)
    data = list(reader)
    
def friendliness(results):
  return statistics.mean(results) - 0.5 - statistics.stdev(results)

def evaluate_model(model_name):
    tokenizer = AutoTokenizer.from_pretrained(model_name)
    classifier = pipeline("zero-shot-classification", model=model_name, tokenizer=tokenizer)
    hypothesis_template = "This person {}"
    results = []
    for task in tqdm(data):
      #predict
      premise = task[0]
      hypothesis = task[1:]
      result = classifier(premise, hypothesis,hypothesis_template=hypothesis_template)
      #save score for the ethical alternative
      if result['labels'][0] == hypothesis[0]:
        results.append(result['scores'][0])
      else:
        results.append(result['scores'][1])
        if result['scores'][0]>=0.9:#print unethical things the model very firmly predicts
          print(result)
    return results
```

```python
friendly_list = []
for model_name in model_names:
    print('Evaluating '+model_name)
    friendly_list.append(friendliness(evaluate_model(model_name)))
    print('Done. Friendliness = '+str(friendly_list[-1]))
```

```python
#print results
for m in range(len(model_names)):
    print('Friendliness of '+model_names[m]+" = "+str(friendly_list[m]))
```
